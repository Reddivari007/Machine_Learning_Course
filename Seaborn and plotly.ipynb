```python
import seaborn as sns
import matplotlib.pyplot as plt



tips=sns.load_dataset('tips')  #loading dataset

print(tips.head(20))
fig,ax1 = plt.subplots(figsize=(10,6))

sns.histplot(data=tips,x='total_bill',kde=True,ax=ax1,bins=5)

ax1.set_title('Distribution of Total Bill')
ax1.set_xlabel('Total Bill')
ax1.set_ylabel('Frequency')

plt.show()
```

```python
# rug plot -> displays each inividual data point as a small tick mark

# shows density -> more lines -> more data points in that region

import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset('tips')

print(tips.head())

fig,ax= plt.subplots(figsize=(15,6))

sns.histplot(data=tips,x='total_bill',kde=True,ax=ax)
sns.rugplot(data=tips,x='total_bill',ax=ax,alpha=0.5,color='red')

ax.set_title('Rug Plot of Total Bill')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset('tips')

print(tips.head())

fig,ax= plt.subplots(figsize=(15,6))

sns.barplot(data=tips,x='day',y='total_bill',ax=ax)
ax.set_title('Bar Plot')
ax.set_xlabel('Day of the week')
ax.set_ylabel('Total bill')

plt.show()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset('tips')

print(tips.head())

fig,ax= plt.subplots(figsize=(15,6))

sns.boxplot(data=tips,x='day',y='total_bill',ax=ax)
sns.violinplot(data=tips,x='day',y='total_bill',ax=ax)
ax.set_title('Violin Plot')
ax.set_xlabel('Day of the week')
ax.set_ylabel('Total bill')

plt.show()
```

```python
import plotly.express as px


iris=px.data.iris()

fig = px.scatter_3d(
    iris,
    x="sepal_length",
    y='sepal_width',
    z='petal_length',
    color='species'
)

# customize 3D scene

fig.update_layout(


                  scene=dict(
                      xaxis_title='Sepal Length',
                      yaxis_title='Sepal Width',
                      zaxis_title='Petal Length'
                  ),
                  width=800,
                  height=700
)


fig.show()
```
