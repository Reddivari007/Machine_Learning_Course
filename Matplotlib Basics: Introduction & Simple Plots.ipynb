```python
# line plot


# basic line plot


import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-10,10,100)
# print(x)

y=np.sin(x)
# print(x,y)

fig, ax = plt.subplots(figsize=(8,5))
ax.plot(x,y,color='steelblue',linewidth=2)

ax.set_title('sine_wave')
ax.set_xlabel('X')
ax.set_ylabel('sin(x)')
ax.grid(True,alpha=0.5)

plt.show()
```

```python
# line plot


# basic line plot


import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-10,10,100)
# print(x)

y=np.cos(x)
# print(x,y)

fig, ax = plt.subplots(figsize=(8,5))
ax.plot(x,y,color='steelblue',linewidth=2)

ax.set_title('cosine_wave')
ax.set_xlabel('X')
ax.set_ylabel('cos(x)')
ax.grid(True,alpha=0.5)

plt.show()
```

```python
# multiple lines + legend

x=np.linspace(0,10,200)


fig, ax = plt.subplots(figsize=(10,5))

ax.plot(x,np.sin(x),label='sin(x)')
ax.plot(x,np.cos(x),label='cos(x)')

ax.set_title('sine vs cosine')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.grid(True,alpha=0.5)

ax.legend()

plt.show()
```

```python
# line styles , markers


x=np.arange(1,6)
y=x
print(x,y)

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x,y, linestyle='--', marker='o',label='dashed line')


ax.set_title('line styles')
ax.legend()
ax.grid(True,alpha=0.3)
plt.show()
```

```python
# basic scatter plot

# plot of points that are scattered

import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)
x=np.random.randn(100)
y=np.random.randn(100)
# plt.scatter(x,y)
# plt.show()

print(x,y)

fig, ax= plt.subplots(figsize=(7,5))

ax.scatter(x,y,alpha=1)

ax.set_title('scatter plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.grid(True,alpha=0.3)

plt.show()
```

```python
# bar chart  -> vertical bar chart and horizontal bar chart

# bar -> vertical bar chart
import matplotlib.pyplot as plt

categories=['A','B','C','D','E']

values = [25,40,30,55,45]

fig , ax = plt.subplots(figsize=(10,5))

ax.bar(categories,values,color='steelblue',alpha=0.8)

ax.set_title('Simple Bar Chart')

ax.set_xlabel('Categories')

ax.set_ylabel('Value')

ax.grid(axis='y',alpha=0.3)
plt.show()
```

```python

# barh -> horizontal bar chart
import matplotlib.pyplot as plt

categories=['A','B','C','D','E']

values = [25,40,30,55,45]

fig , ax = plt.subplots(figsize=(10,5))

ax.barh(categories,values,color='steelblue',alpha=0.8)

ax.set_title('Simple Bar Chart')

ax.set_xlabel('Value')

ax.set_ylabel('Categories')

ax.grid(axis='x',alpha=0.3)
plt.show()
```
