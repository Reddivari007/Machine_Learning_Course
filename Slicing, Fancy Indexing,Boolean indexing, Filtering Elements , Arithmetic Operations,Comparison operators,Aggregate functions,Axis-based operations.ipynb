```python
# Fancy indexing -> uses a list or array of indices to access multiple elements simultaneously

arr1=np.array([10,20,30,40])

print(arr1)
```

```python
# access 10,30,40

print(arr1[0])
print(arr1[2])
print(arr1[3])
```

```python
print(arr1[[0,2,3]])
```

```python
# boolean indexing -> allows selection of elements based on a condition that returns True/False

print(arr1)
```

```python
# lets say we want to see the values which are greater than 20

print(arr1[arr1>20])
# [10 20 30 40]
# print(arr1[40>20])
# print(arr1[True])
```

```python
print(arr1[arr1>10])
```

```python
print(arr1[(arr1>10) & (arr1<30)])   # | -> or

# T | T -> T
# T | F -> T
# F | T -> T
# F | F -> F
# T & T -> T
# T & F -> F
# F & T -> F
# F & F -> F
```

```python
print(arr1[(arr1%3)==0])
```

```python
# Arithmetic Operations

# -> numpy supports element wise arithmetic operations between arrays of same shape or compatible shapes

a=np.array([1,2,3])
b=np.array([4,5,6])
print(a+b)


```

```python
print(a*b)
```

```python
# broadcasting rules  -> allows arithmetic operations on arrays of different shapes by automatically
# expanding the smaller array

a=np.array([[1,2,3],[4,5,6]])
print(a)

print(a+10)
```

```python
# aggregate functions

print(a)
```

```python
print(a.sum()) #returns sum of all the elements
```

```python
print(a.mean())
```

```python
# axis=0 -> sums up along columns
# axis=1 -> sums up along rows

print(a)
print(a.sum(axis=0))
print(a.sum(axis=1))
```

```python
print(a.mean(axis=0))
print(a.mean(axis=1))
```

```python
print(a.max())
```

```python
print(a.min())
```

```python
print(a.max(axis=0))
print(a.max(axis=1))
```

```python
print(a.min(axis=0))
print(a.min(axis=1))
```
