```python
# sorting -> to arrange the data in asc or desc order
# sort_values() -> used to sort a DataFrame on values in one or more columns

```

```python
import pandas as pd
df = pd.DataFrame( {
        'Name':['A','B','C','D'],
         'Age': [10,12,13,10]
})

```

```python
print(df)
```

```python
print(df.sort_values('Age'))  #by default in ascending order of age column
```

```python
print(df.sort_values('Age',ascending=False))  #by default in ascending order of age column
```

```python
print(df.sort_index(ascending=False))
```

```python
# rank()

# -> assign ranks to a column

# smallest value gets rank 1 ..ties get average rank by default

df['Rank']= df['Age'].rank()
print(df)



# ties get average rank -> since A and D has same marks ..they block both 1st and 2nd positions..
# and average of the blocked positions is assigned as ranks
```

```python
# conditional filters

# filter and then query

# age>10 and name!=C

print(df[(df['Age']>10) & (df['Name']!='C')])

# T and T ? T
# T and F ? False
# F and T ? False
# F and F is False
```

```python
print(df[(df['Age']>10) | (df['Name']!='C')])
# t or F is T
# F or T is T
# T or T is T
# F or F is F
```

```python
# grouping and aggregation

# groupby() -> a function that is used to split data into groups based on some criterio


# what is average salary of IT dept -> group all the employees who are having dept='id' ..take avg of all the rows
```

```python
data = {'Category': ['A', 'A', 'B', 'B', 'C'],
        'Values': [10, 20, 30, 40, 50]}

df=pd.DataFrame(data)

grouped =df.groupby('Category')
print(grouped)
print(grouped.sum())
print(grouped.mean())
print(grouped.max())
print(grouped.min())
print(grouped.count())
```

```python
# agg() -> helps us apply multiple aggregation functions on a column at once

print(df.groupby('Category')['Values'].agg(['sum','mean','count']))
```

```python
# transform() -> returns a result with the same shape as the input.. broadcasting the values back

print(df.groupby('Category')['Values'].transform('mean'))
```

```python
print(df)
```

```python
# apply() -> allows applying a function accross groups and can return different shapes

print(df.groupby('Category')['Values'].apply(lambda x: x.max()))
```

```python
# pd.concat()

df1 = pd.DataFrame({'A': [1,2], 'B':[3,4]})
df2 = pd.DataFrame({'A': [5,6], 'B':[7,8]})
```

```python
result=pd.concat([df1,df2])
print(df1)
print(df2)
print(result)
```

```python
# merge() -> similar to  join operation we have in SQL

df1 = pd.DataFrame({'ID':[1,2,3], 'Name':['A','B','C']})
df2 = pd.DataFrame({'ID':[1,2,4], 'Score':[90,80,70]})

# now we want to have ID,Name and score in the output
# 1,A,90
# 2,B,80

print(pd.merge(df1,df2,on='ID',how='inner'))
```

```python
print(pd.merge(df1,df2,on='ID',how='left'))
```

```python
print(pd.merge(df1,df2,on='ID',how='right'))
```

```python
print(pd.merge(df1,df2,on='ID',how='outer'))
```
