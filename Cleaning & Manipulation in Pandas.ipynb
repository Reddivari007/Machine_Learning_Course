```python
# 20 -> 5 columns -> 1 column
data = {
    "A": [1,2,None,4],
    "B":[10,None,30,40]
}

df = pd.DataFrame(data)
print(df)
```

```python
# check missing values  isnull()

print(df.isnull())
```

```python
# dropping missing rows   -> droppingna

print(df.dropna()) #drop missing rows
```

```python
# replace missing values with 0

print(df.fillna(0))
```

```python
# replace values

df=pd.DataFrame({'Grade':['A','B','C']})

# Excellent,good and average
print(df)
print(df.replace({'A':"Excellent",'B':'good','C':'average'}))
```

```python
# renaming of columns

df=pd.DataFrame({'col1':[1,2],'col2':[3,4]})
print(df.columns)

df.rename(columns={'col1':'column1','col2':'column2'},inplace=True)

# print(df.rename(columns={'col1':'column1','col2':'column2'}))
print(df)
```

```python
# add,update,delete column

df=pd.DataFrame({

                'Name':['A','B','C'],
                'Age':[10,20,30],
                'City':['X','Y','Z']
})

print(df)
```

```python
# adding a new column
df['salary']=[1000,1500,1200]
```

```python
print(df)
```

```python
# add columns with some calculations
df['age_after_5'] =df['Age']+5
print(df)
```

```python
# adding a row

df.loc[3]=['D',12,'P',1300,17]
```

```python
print(df)
```

```python
# updating entire column
df['City']=df['City'].str.lower()

print(df)
```

```python
df.loc[3,'Age']=13
```

```python
print(df)
```

```python
# df.loc[3,'Age']=13

# how to update using a condition ..increase salary by 10 oercent of all the people
# having age>15

df.loc[df['Age']>15,'salary']*=1.1
# a*=1.1  a=a*1.1
print(df)

```

```python
print(df.loc[df['Age']>15,'salary'])
```

```python
# updating multiple columns at once

df.loc[0,['Age','City']]=[11,'hyd']
print(df)
```

```python
# deletion

# deleting a column
df.drop('age_after_5',axis=1,inplace=True)
print(df)
```

```python
# deleting multiple columns
df.drop(['Age','salary'],axis=1,inplace=True)
print(df)
```

```python
# deleting a row

df.drop(1,axis=0,inplace=True)
print(df)
```

```python
# resetting indexes after deletion
df.reset_index(drop=True,inplace=True)
print(df)
```

```python
df=pd.DataFrame({'Numbers':[1,2,3,4]})

# map() works in Series
df['Squares']=df['Numbers'].map(lambda x:x**2)
print(df)
```

```python
df['Cubes']=df['Numbers'].map(lambda x:x**3)
print(df)
```

```python
# apply() works on series
df['Cubes']=df['Numbers'].apply(lambda x:x**3)
print(df)
```

```python
# applymap()

df2=df[['Squares','Cubes']]
print(df2)

```

```python
df2=df2.applymap(lambda x:x*10)
print(df2)
```
