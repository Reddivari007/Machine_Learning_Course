```python
import numpy as np
import time
import sys

size=100000

python_list = range(size)
numpy_array = np.arange(size)

```

```python
print("Performance comparision between Numpy Arrays and Python Lists")

print("Element wise addition( adding 5 to each element)")

start = time.time()
result_list = [x+5 for x in python_list]
end=time.time()
list_time = end-start

start = time.time()
result_array = numpy_array + 5
end = time.time()
numpy_time = end - start

print(list_time)
print(numpy_time)
```

```python

print("Element wise multiplication( multiplying 5 to each element)")

start = time.time()
result_list = [x*5 for x in python_list]
end=time.time()
list_time = end-start

start = time.time()
result_array = numpy_array * 5
end = time.time()
numpy_time = end - start

print(list_time)
print(numpy_time)
```

```python
# eye(n) -> creates a 2D identity-like matrix of size nxn, with 1s on the diagonal and 0s elsewhere

eye_mat = np.eye(3) #-> creates a 3x3 identity matrix
print(eye_mat)
```

```python
# np.identity(n) -> creates a 2D identity matrix of nxn dimensions

identity_mat = np.identity(3)
print(identity_mat)
```

```python
# np.zeros(shape) -> creates an array filled with zeros, given the desired shape

import numpy as np
zeros_array = np.zeros((3,5))
print(zeros_array)
```

```python
# np.ones(shape) -> creates and array filled with ones, given the desired shape

ones_array=np.ones((3,5))
print(ones_array)
```

```python
# name,age,gpa,graduated

student_dtype = np.dtype([
    ('name','U20'), #unicode string with 20 maximum chars
    ('age','i4'), #32-bit integer
    ('gpa','f8'),  #64 bit -floating point number
    ('graduated','?') #boolean
])

# name,age,gpa,graduated
# Eswar, 30,8.8,yes
# Sai, 20,yes, 9.8
```
