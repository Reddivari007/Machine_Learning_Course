```python
# indexing,selection,cleaning and manipulation in pandas

import pandas as pd

data={
    "Name":['A','B','C','D'],
    "age":[10,12,14,11],
    "city":['hyd','blr','del','vjy']

  }

df= pd.DataFrame(data)
print(df)
```

```python
print(df.describe())
```

```python
print(df['age'].unique())
print(df['age'].nunique())
print(df['age'].value_counts())
```

```python
# selecting a single column

print(df['Name'])
```

```python
# select multiple columns

print(df[["Name",'age']])
```

```python
# df.loc()

print(df.loc[3])
```
